# PART A: RNN - Character-Level Language Model (10 min)

**Use Case:** Predict the next character for autocomplete.

**Dataset:** Tiny Shakespeare

```python
# Load Tiny Shakespeare dataset
from datasets import load_dataset

shakespeare = load_dataset("tiny_shakespeare", split="train")
text = shakespeare['text'][0][:10000]  # Use first 10K chars for speed

print(f"Text length: {len(text)} characters")
print(f"Sample: {text[:200]}")
```

```python
# Create character vocabulary
chars = sorted(list(set(text)))
vocab_size = len(chars)
char_to_idx = {ch: i for i, ch in enumerate(chars)}
idx_to_char = {i: ch for i, ch in enumerate(chars)}

print(f"Vocabulary size: {vocab_size}")
print(f"Characters: {''.join(chars[:30])}...")
```

```python
# Prepare sequences
seq_length = 30
X, y = [], []

for i in range(len(text) - seq_length):
    X.append([char_to_idx[c] for c in text[i:i+seq_length]])
    y.append(char_to_idx[text[i+seq_length]])

X = torch.tensor(X, dtype=torch.long)
y = torch.tensor(y, dtype=torch.long)

print(f"Sequences: {X.shape[0]}, Sequence length: {seq_length}")
```

```python
# Simple RNN model
class CharRNN(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.rnn = nn.RNN(embed_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)
    
    def forward(self, x):
        x = self.embedding(x)
        out, _ = self.rnn(x)
        out = self.fc(out[:, -1, :])  # Last timestep
        return out

# Create model
rnn_model = CharRNN(vocab_size, embed_dim=32, hidden_dim=64).to(device)
print(f"RNN Parameters: {sum(p.numel() for p in rnn_model.parameters()):,}")
```

### Exercise A.1: Train the RNN

```python
# Complete the training loop

# Hyperparameters
batch_size = 128
epochs = 5
learning_rate = 0.005  # 0.001-0.01 range; 0.005 gives a good speed/stability trade-off here

# DataLoader
dataset = TensorDataset(X[:5000], y[:5000])  # Use subset for speed
loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(rnn_model.parameters(), lr=learning_rate)

# Training loop
losses = []
for epoch in range(epochs):
    rnn_model.train()
    total_loss = 0
    for batch_X, batch_y in loader:
        batch_X, batch_y = batch_X.to(device), batch_y.to(device)

        # 1. Zero gradients
        optimizer.zero_grad()
        # 2. Forward pass
        output = rnn_model(batch_X)
        # 3. Compute loss
        loss = criterion(output, batch_y)
        # 4. Backward pass
        loss.backward()
        # 5. Update weights
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(loader)
    losses.append(avg_loss)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}")
```

```python
# Generate text
def generate_text(model, start_str, length=100):
    model.eval()
    chars_generated = list(start_str)
    input_seq = [char_to_idx.get(c, 0) for c in start_str[-seq_length:]]
    
    with torch.no_grad():
        for _ in range(length):
            x = torch.tensor([input_seq[-seq_length:]]).to(device)
            output = model(x)
            pred_idx = torch.argmax(output, dim=1).item()
            chars_generated.append(idx_to_char[pred_idx])
            input_seq.append(pred_idx)
    
    return ''.join(chars_generated)

# Test generation
print("Generated text:")
print(generate_text(rnn_model, "To be or not", length=100))
```

# PART B: LSTM - Sentiment Analysis 

**Use Case:** Classify movie review sentiment.

**Dataset:** IMDB Reviews

```python
# Load IMDB dataset
imdb = load_dataset("imdb")

# Small sample for quick training
train_texts = imdb['train']['text'][:1000]
train_labels = imdb['train']['label'][:1000]
test_texts = imdb['test']['text'][:200]
test_labels = imdb['test']['label'][:200]

print(f"Train: {len(train_texts)}, Test: {len(test_texts)}")
```

```python
# Simple tokenization and vocabulary
from collections import Counter
import re

def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())[:100]  # Max 100 tokens

# Build vocabulary from training data
all_tokens = [tok for text in train_texts for tok in tokenize(text)]
vocab = {word: idx+2 for idx, (word, _) in enumerate(Counter(all_tokens).most_common(5000))}
vocab['<PAD>'] = 0
vocab['<UNK>'] = 1

print(f"Vocabulary size: {len(vocab)}")
```

```python
# Encode texts
def encode_text(text, max_len=100):
    tokens = tokenize(text)
    encoded = [vocab.get(t, 1) for t in tokens]  # 1 = UNK
    padded = encoded[:max_len] + [0] * (max_len - len(encoded))
    return padded[:max_len]

X_train = torch.tensor([encode_text(t) for t in train_texts])
y_train = torch.tensor(train_labels)
X_test = torch.tensor([encode_text(t) for t in test_texts])
y_test = torch.tensor(test_labels)

print(f"Train shape: {X_train.shape}")
```

### Exercise B.1: Complete the LSTM Model

```python
# Complete the LSTM classifier

class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, num_classes):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)

        # Define LSTM layer
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)

        self.fc = nn.Linear(hidden_dim, num_classes)
        self.dropout = nn.Dropout(0.3)

    def forward(self, x):
        x = self.embedding(x)
        x = self.dropout(x)

        # Pass through LSTM and get final hidden state
        lstm_out, (hidden, cell) = self.lstm(x)

        out = self.fc(hidden[-1])  # Use last layer's final hidden state
        return out

# Create model
lstm_model = LSTMClassifier(
    vocab_size=len(vocab),
    embed_dim=64,
    hidden_dim=64,
    num_classes=2
).to(device)

print(f"LSTM Parameters: {sum(p.numel() for p in lstm_model.parameters()):,}")
```

```python
# Quick training
train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(lstm_model.parameters(), lr=0.001)

# Train for 3 epochs
for epoch in range(3):
    lstm_model.train()
    total_loss = 0
    for batch_X, batch_y in train_loader:
        batch_X, batch_y = batch_X.to(device), batch_y.to(device)
        optimizer.zero_grad()
        output = lstm_model(batch_X)
        loss = criterion(output, batch_y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss/len(train_loader):.4f}")

# Evaluate
lstm_model.eval()
with torch.no_grad():
    test_output = lstm_model(X_test.to(device))
    preds = torch.argmax(test_output, dim=1).cpu()
    acc = (preds == y_test).float().mean()
    print(f"\nTest Accuracy: {acc:.4f}")
```

# PART C: GRU - News Classification

**Use Case:** Classify news articles by topic.

**Why GRU?** Fewer parameters than LSTM, faster training.

```python
# Load AG News
ag_news = load_dataset("ag_news")
ag_train = ag_news['train'].shuffle(seed=42).select(range(2000))
ag_test = ag_news['test'].shuffle(seed=42).select(range(500))

ag_labels = {0: 'World', 1: 'Sports', 2: 'Business', 3: 'Sci/Tech'}
print(f"Classes: {list(ag_labels.values())}")
```

### Exercise C.1: Build GRU Classifier

```python
# Create a GRU classifier (similar to LSTM but using nn.GRU)

class GRUClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, num_classes):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)

        # Define GRU layer
        self.gru = nn.GRU(embed_dim, hidden_dim, batch_first=True)

        self.fc = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        x = self.embedding(x)

        # GRU forward pass
        # Note: GRU returns (output, hidden) - no cell state unlike LSTM
        gru_out, hidden = self.gru(x)

        out = self.fc(hidden[-1])  # Use last layer's final hidden state
        return out

# Build vocabulary and encode (reuse tokenize function)
ag_tokens = [tok for item in ag_train for tok in tokenize(item['text'])]
ag_vocab = {word: idx+2 for idx, (word, _) in enumerate(Counter(ag_tokens).most_common(5000))}
ag_vocab['<PAD>'] = 0
ag_vocab['<UNK>'] = 1

def encode_ag(text, vocab, max_len=50):
    tokens = tokenize(text)
    encoded = [vocab.get(t, 1) for t in tokens]
    return (encoded[:max_len] + [0] * max_len)[:max_len]

X_ag_train = torch.tensor([encode_ag(item['text'], ag_vocab) for item in ag_train])
y_ag_train = torch.tensor([item['label'] for item in ag_train])
X_ag_test = torch.tensor([encode_ag(item['text'], ag_vocab) for item in ag_test])
y_ag_test = torch.tensor([item['label'] for item in ag_test])

print(f"AG News - Train: {X_ag_train.shape}, Test: {X_ag_test.shape}")
```

```python
# Create and train GRU model
gru_model = GRUClassifier(
    vocab_size=len(ag_vocab),
    embed_dim=64,
    hidden_dim=64,
    num_classes=4
).to(device)

print(f"GRU Parameters: {sum(p.numel() for p in gru_model.parameters()):,}")
print(f"(Compare to LSTM: GRU has fewer parameters!)\n")

# Train
ag_train_ds = TensorDataset(X_ag_train, y_ag_train)
ag_train_loader = DataLoader(ag_train_ds, batch_size=32, shuffle=True)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(gru_model.parameters(), lr=0.001)

for epoch in range(3):
    gru_model.train()
    total_loss = 0
    for batch_X, batch_y in ag_train_loader:
        batch_X, batch_y = batch_X.to(device), batch_y.to(device)
        optimizer.zero_grad()
        output = gru_model(batch_X)
        loss = criterion(output, batch_y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss/len(ag_train_loader):.4f}")

# Evaluate
gru_model.eval()
with torch.no_grad():
    ag_out = gru_model(X_ag_test.to(device))
    ag_preds = torch.argmax(ag_out, dim=1).cpu()
    ag_acc = (ag_preds == y_ag_test).float().mean()
    print(f"\nTest Accuracy: {ag_acc:.4f}")
```

```python
import requests

def query_mistral(prompt, max_tokens=150):
    """Query Mistral API."""
    url = "https://api.mistral.ai/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {MISTRAL_API_KEY}",
        "Content-Type": "application/json"
    }
    data = {
        "model": "mistral-small-latest",
        "messages": [{"role": "user", "content": prompt}],
        "max_tokens": max_tokens
    }
    
    response = requests.post(url, headers=headers, json=data)
    if response.status_code == 200:
        return response.json()['choices'][0]['message']['content']
    else:
        return f"Error: {response.status_code} - {response.text}"

# Test (only if API key is set)
if MISTRAL_API_KEY != "___":
    response = query_mistral("What is NLP in one sentence?")
    print(f"Mistral: {response}")
else:
    print("Please set your MISTRAL_API_KEY above.")
```

*(The exact numbers print from each model'''s `sum(p.numel() ...)` line. The embedding table alone is `5002 x 64 = 320K` params, so it dominates both totals — the LSTM vs GRU difference lives entirely in the recurrent cell.)*

**Why does LSTM have more parameters than GRU?**

The difference is in the recurrent cell, not the embedding. An LSTM has **three gates plus a candidate cell state** — the input, forget, and output gates, and the cell-state update — so it learns **4** weight blocks per step. A GRU merges these into **two gates** (update and reset) plus one candidate hidden state, giving **3** weight blocks. Each block is about `hidden x (embed + hidden)` weights plus biases, so the LSTM carries ~4/3 the recurrent parameters of the GRU: with `hidden = 64` that is ~33K (LSTM) vs ~25K (GRU), an ~8K gap — exactly the difference between the ~353K and ~345K totals. Fewer gates is also why the GRU trains faster and uses less memory, at the cost of slightly less expressive gating.


### Question 2: RNN vs LSTM for Long Sequences

**Why would LSTM perform better than vanilla RNN for sentiment analysis on long reviews?** Explain the vanishing gradient problem.

**YOUR ANSWER:**

A vanilla RNN updates its hidden state at every timestep by repeatedly multiplying by the same recurrent weight matrix. During backpropagation-through-time, the gradient flowing from a late token back to an early token is a **product of many such factors**. If those factors are consistently below 1 the gradient shrinks exponentially toward zero (**vanishing gradient**); if above 1 it blows up (exploding gradient). The practical consequence: for a long movie review, the RNN's gradient signal from the final classification barely reaches the words near the start, so the model effectively **forgets long-range context** and can only learn short-range patterns.

The LSTM fixes this with its **cell state and gating**. The cell state acts as a mostly-linear "conveyor belt" that carries information across timesteps with only additive updates, and the **forget gate** lets the network decide what to keep or discard rather than overwriting everything each step. Because information can flow along the cell state with gradients close to 1, the error signal survives across many timesteps. This is why the LSTM captures the long-range dependencies in a long review — e.g., a negation early on ("not good at all...") still influencing the final sentiment prediction — whereas the vanilla RNN loses that signal.


### Question 3: Traditional Models vs LLMs

Based on your experiments:
1. **What can LLMs do that LSTM/GRU cannot?**
2. **What are the disadvantages of using LLM APIs?** (Think: cost, latency, privacy)
3. **When would you choose a traditional model over an LLM?**

**YOUR ANSWER:**

1. **LLM advantages (what LSTM/GRU cannot easily do):** The LLM works **zero-shot** — it did sentiment classification and one-sentence summarization with no task-specific training and no labelled dataset, just a prompt. Our LSTM was trained only to output a 2-class sentiment label; it cannot summarize, answer open questions, translate, or reason. The LLM also brings broad world knowledge and handles arbitrary instructions, whereas the LSTM/GRU only know the narrow task and vocabulary they were fit on.

2. **LLM disadvantages:** **Cost** — every API call is billed per token, so at scale it is far more expensive than running a tiny local LSTM. **Latency** — a network round-trip to the API (hundreds of ms to seconds) versus a near-instant local forward pass. **Privacy / data control** — text is sent to a third-party server, which is unacceptable for sensitive or regulated data. Add **dependency and reliability** (rate limits, outages, model changes) and **non-determinism** (outputs can vary between calls).

3. **When to choose a traditional model:** When the task is **narrow, high-volume, and latency- or cost-sensitive** (e.g., classifying millions of reviews), when data **must stay on-premise** for privacy/compliance, when you need **deterministic, reproducible** outputs, or when you must run **offline / on constrained hardware**. A small LSTM/GRU is cheap, fast, private, and — once trained on enough in-domain data — can match or beat an LLM on that one specific task.
